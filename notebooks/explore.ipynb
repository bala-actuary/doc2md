# doc2md exploration

Use this notebook to test conversions on real documents and tune OCR settings before relying on the CLI.

All real logic lives in `doc2md.converter.convert` — this notebook is a thin harness for inspection.

Replace the placeholder paths below with real documents on your machine.

```python
from pathlib import Path
from doc2md import convert
```

## 1. Text PDF (no OCR needed)

Modern Word/LaTeX-exported PDFs typically have extractable text plus embedded diagrams. Docling handles these natively — OCR does not fire.

```python
source_pdf = r"PATH\TO\YOUR\document.pdf"

out = convert(
    source_pdf,
    r"..\outputs\document.md",
)
print("Wrote:", out)
print("Size:", out.stat().st_size, "bytes")
```

```python
# Preview first 2000 chars
print(out.read_text(encoding="utf-8")[:2000])
```

## 2. Scanned PDF with forced OCR

For image-only PDFs (scans), force Tesseract via `ocr_langs`. Use ISO 639-2 codes: `eng`, `tam`, `fra`, `deu`, etc. Tesseract needs the matching language pack installed on the system.

```python
scanned_pdf = r"PATH\TO\YOUR\scanned_document.pdf"

out = convert(
    scanned_pdf,
    r"..\outputs\scanned_document.md",
    ocr_langs=["eng"],
)
print("Wrote:", out)
```

```python
# Inspect OCR output
print(out.read_text(encoding="utf-8")[:3000])
```
